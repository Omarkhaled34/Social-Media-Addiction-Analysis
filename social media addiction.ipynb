```python
import pandas as pd
```

```python
df=pd.read_csv("D:/files/archive_dataset/Students Social Media Addiction.csv")
```

```python
df.head(5)
```

```python
print(f"the number of duplicates is {df.duplicated().sum()}")
```

```python
df.shape
```

```python
avg_age=df['Age'].mean()
avg_age
```

```python
filt=(df['Age']<avg_age) & (df['Mental_Health_Score']<=5) 
print("people with low mental health under the age 20")
df[filt]['Gender'].value_counts()
```

```python
df['Relationship_Status'].unique()
```

```python
print(f"the least addicted score: {df['Addicted_Score'].min()}")
print(f"the most addicted score: {df['Addicted_Score'].max()}")
```

```python
filt=df['Addicted_Score']>5
print("Relationship status distribution for people with high addiction score:")
df[filt]['Relationship_Status'].value_counts()
```

```python
print("average nightly sleep hours across different relationship categories")
df.groupby('Relationship_Status')['Sleep_Hours_Per_Night'].mean()
```

```python
df['Most_Used_Platform'].unique()
```

```python
print("the most used apps by avg daily use:")
df.groupby("Most_Used_Platform")["Avg_Daily_Usage_Hours"].agg(['count','mean'])\
.sort_values(by="count",ascending=False)
```

```python
print("the most academic level addicted to social media by avg daily use and mental health:")
df.groupby(["Academic_Level"])[["Avg_Daily_Usage_Hours",'Mental_Health_Score']]\
    .mean().sort_values(by="Avg_Daily_Usage_Hours",ascending=False)
```

```python
df['Conflicts_Over_Social_Media'].unique()
```

```python
print("the corelation between Daily_Usage_Hours and Conflicts_Over_Social_Media: ")
print(df['Avg_Daily_Usage_Hours'].corr(df['Conflicts_Over_Social_Media']))
```

```python
print("the corelation between Daily_Usage_Hours and mental health: ")
print(df['Avg_Daily_Usage_Hours'].corr(df['Mental_Health_Score']))
```

```python
print('Relationship Between Daily Usage Hours and Academic Performance:')
df.groupby('Affects_Academic_Performance')['Avg_Daily_Usage_Hours'].mean()
```

```python
print('the corelation between Sleep_Hours_Per_Night and mental health:')
print(df['Sleep_Hours_Per_Night'].corr(df['Mental_Health_Score']))
```

```python
print("the top 5 platforms with the lower mental health users:")
df.groupby('Most_Used_Platform')['Mental_Health_Score'].mean()\
    .sort_values(ascending=True).head(5)
```
